```python
def clean_string(obj):
    new_string=re.sub('[^a-zA-Z0-9]',' ', re.sub("[\(\[].*?[\)\]]", "", obj))
    cleaned_string=re.sub('\s+',' ',new_string)
    return cleaned_string
```

```python
# elements in genres column are strings
# convert them to list
import ast
def getValues(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i)
    return L

kdramas['genres'] = kdramas['genres'].apply(getValues)
```

```python
def getValues(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
kdramas['keywords'] = kdramas['keywords'].apply(getValues)
```

```python
def get10Casts(obj):
    L = []
    count = 0 
    for i in ast.literal_eval(obj):
        if (count < 10) :
            L.append(i['name'])
            count+=1
        else:
            break
    return L
```

```python
kdramas['cast'] = kdramas['cast'].apply(get10Casts)
```

```python
def fetchDirectorAndWriter(obj):
    L = []
    for i in ast.literal_eval(obj):
        if (i['job']=='Director' or i['job']=='Writer' or i['job']=='Screenplay') :
            L.append(i['name'])
    return L
```

```python
kdramas['crew'] = kdramas['crew'].apply(fetchDirectorAndWriter)
```

```python
kdramas['overview'] = kdramas['overview'].apply(lambda x:x.split())
```

```python
kdramas.head()
```

```python
# remove spaces between a genre name, overview, cast name, crew name
kdramas['genres'] = kdramas['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
kdramas['overview'] = kdramas['overview'].apply(lambda x:[i.replace(" ", "") for i in x])
kdramas['cast'] = kdramas['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
kdramas['crew'] = kdramas['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
kdramas.head()
```

```python
kdramas['tags'] = kdramas['overview'] + kdramas['genres'] + kdramas['keywords'] + kdramas['cast'] + kdramas['crew']
kdramas.head()
```

```python
kdramas['tags'] = kdramas['tags'].apply(lambda x:" ".join(x))
```

```python
kdramas['tags'] = kdramas['tags'].apply(lambda x:x.lower())
```

```python
kdramas_df = kdramas[['id', 'name', 'tags']]
kdramas_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english', max_features=5000)
```

```python
vectors = cv.fit_transform(kdramas_df['tags']).toarray()
vectors
```

```python
cv.get_feature_names_out()
```

```python
# since there are alot of similar words
# like accept, accepts, accepted
# apply stemming on words

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
kdramas_df['tags'] = kdramas_df['tags'].apply(stem)
```

```python
kdramas_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english', max_features=3000)
vectors = cv.fit_transform(kdramas_df['tags']).toarray()
vectors
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity.shape
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(kdrama):
    kdrama_index = kdramas_df[kdramas_df['title'] == kdrama].index[0]
    distances = similarity[kdrama_index]
    kdramas_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    for i in kdramas_list:
        print(kdramas_df.iloc[i[0]].title)
```

```python
recommend('Crash Landing on You')
```

```python
def recommend(kdrama):
    kdrama_index = kdramas_df[kdramas_df['title'] == kdrama].index[0]
    distances = similarity[kdrama_index]
    kdramas_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i, (index, score) in enumerate(kdramas_list, 1):
        recommended_kdrama = kdramas_df.iloc[index].title
        print(f"{i}. {recommended_kdrama} (Cosine Similarity: {score})")
```
